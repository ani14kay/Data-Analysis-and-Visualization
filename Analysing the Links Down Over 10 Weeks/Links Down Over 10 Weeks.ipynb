
# Analysing & Visualizing Number of Links Down Weekly


### [ 1 ] Importing Appropriate Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

### [ 2 ] Uploading the Downloaded Dataset and Displaying Contents

```python
dts = pd.read_csv("Num_of_links_down_per_week.csv")
dts
```

### [ 3] Viewing the list of columns in the dataset

```python
dts.columns
```

### [ 4 ] Copying data to new data frame so that original remains unaffected 

```python
df = dts.copy()
```

### [ 5 ] Looking through the Data Frame

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

##### [ 5.1 ] Splitting Link Column into its Sub-Parts

```python
df[['link from','device from','device to','link to']]=df['link'].str.split(':',expand=True)
df
```

##### [ 5.2 ] Scenario - 1 : Direct Failures, links down before Config Establishment

```python
# Destination wasn't reached!
dff = pd.isnull(df['device to'])
df[dff]
```

### [ 6 ] Cleaning dataframe of Un-Established Links

```python
dff = df.dropna(axis = 0, how='any')
```

##### [ 6.1 ] Scenario - 2 : Failures due to other reasons, links down after Config Establishment

```python
# New dataframe dff containing clean data when Destination was reached!
dff
```

### [ 7 ] Setting Display Parameters for Visualization

```python
%matplotlib inline
sns.set_style('darkgrid')
mpl.rcParams['font.size'] = 15
mpl.rcParams['figure.figsize'] = (9, 5)
mpl.rcParams['figure.facecolor'] = '#00000000'
```

### [ 8 ] Exploratory Analysis and Visualization

##### [ 8.2 ] Figuring Out Average Number of Links down Weekly 

```python
plt.figure(figsize=(10,6))
plt.title('Average Links Down Weekly')
sns.barplot(x=dff.index)
```

```python
plt.figure(figsize=(25,6))
plt.title('Links Down Weekly')
sns.countplot(x=dff.index)
```

##### [ 8.3 ] Figuring Out Number of Links down Week-wise Using Different Plots

{ Week - 1 } [ Bargraph ]

```python
plt.figure(figsize=(25,6))
plt.xticks(rotation=70)
sns.barplot(x=dff.index,y=dff.week_1)
plt.xlabel('Index Number')
plt.ylabel('Number of Links Down')
plt.title('Links Down During Week-1')
```

{ Week - 2 } [ Histogram ]

```python
sns.histplot(x=dff.week_2, bins=30, kde=True);
plt.title('Links Down During Week-2')
```

{ Week - 3 } [ Scatter Plot ]

```python
plt.figure(figsize=(20,6))
plt.title('Links Down During Week-3')
sns.scatterplot(x='link', y='week_3', data=dff)
```

{ Week - 4 } [ Kde Plot ]

```python
plt.xlabel('Index Number')
plt.ylabel('Number of Links Down')
plt.title('Links Down During Week-4')
sns.kdeplot(dff['week_4'],shade=True)
```

{ Week - 5 } [ Violin Plot ]

```python
sns.violinplot(x="week_5", data=dff,palette=["yellow"]);
plt.title('Links Down During Week-5')
```

##### [ 8.4 ] Figuring Out Number of Links down Weekly in a Single Graph

{ Weeks - 6 & 7 }

```python
sns.jointplot(x="week_6", y="week_7", data=dff, kind='reg');
plt.title('Links Down During Weeks - 6 & 7', loc='left')
```

{ Weeks - 8, 9 & 10 }

```python
sns.distplot(dff['week_8'])
sns.distplot(dff['week_9'])
sns.distplot(dff['week_10'])
plt.xlabel('Weeks')
plt.title('Links Down During Weeks - 8, 9 & 10')
plt.legend('ENTENT', ncol=2, loc='upper right');
```

##### [ 8.5 ] Graphs Compairing Links Down Over a Period of 10 Weeks

{ Using Pair Plot }

```python
sns.pairplot(dff);
```

{ Using Pair Grid }

```python
f=sns.PairGrid(dff)
f.map(plt.scatter, alpha=0.8)
```

### [ 9 ] Observations

###### --> We observe :
   1) The Trends when Links were down Week-wise using Barplot, Countplot, Histplot, Scatterplot, Kdeplot, Violinplot.
   
   2) The Trends when Links were down when comparing between Weeks using Jointplot and Distplot.
   
   3) The Trends when Links were down when compared throughout the period of 10 weeks, using Pair Plots or Grid Plots.
